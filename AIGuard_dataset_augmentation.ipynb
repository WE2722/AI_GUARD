```python
import os
import cv2
import numpy as np

def calculate_sharpness(image_path):
    """Calculate sharpness using the Laplacian variance method."""
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        return 0  # Handle unreadable images
    return cv2.Laplacian(image, cv2.CV_64F).var()

def find_best_image(folder_path):
    """Find the best image in a folder based on sharpness and resolution."""
    best_score = -1
    best_image = None

    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if file_name.lower().endswith(('.jpg', '.jpeg', '.png')):
            sharpness = calculate_sharpness(file_path)
            resolution = os.path.getsize(file_path)  # Approximation of quality
            score = sharpness * resolution

            if score > best_score:
                best_score = score
                best_image = file_path

    return best_image

def retain_best_images(base_path):
    """Traverse folders and retain only the best image in each."""
    for root, dirs, files in os.walk(base_path):
        images = [f for f in files if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
        if len(images) > 1:
            best_image = find_best_image(root)
            for file_name in images:
                file_path = os.path.join(root, file_name)
                if file_path != best_image:
                    os.remove(file_path)  # Remove non-best images

# Specify the base directory
base_path = r"C:\Users\Wiam\Desktop\AI security alarm system\LFW\lfw-deepfunneled\lfw-deepfunneled"

# Retain only the best image per folder
retain_best_images(base_path)

print("Retained only the best images in each folder.")
```

```python
import cv2
import os
import dlib
from imutils import face_utils

def augment_pose(image_path, output_folder):
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error loading image {image_path}")
        return

    # Load dlib's pre-trained face detector and predictor
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor(r"C:\Users\Wiam\Desktop\AI security alarm system\shape_predictor_68_face_landmarks.dat\shape_predictor_68_face_landmarks.dat")

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)

    if len(faces) == 0:
        print(f"No face detected in {image_path}")
        return

    for face in faces:
        # Get the landmarks
        landmarks = predictor(gray, face)
        landmarks = face_utils.shape_to_np(landmarks)

        # Get the center of the face (use the nose or the center of the face landmarks for the rotation)
        center = (int(landmarks[30][0]), int(landmarks[30][1]))  # Nose is often a good reference point

        # Simulate left and right rotations using affine transformations
        left_matrix = cv2.getRotationMatrix2D(center, 20, 1)  # Rotate 20 degrees left
        right_matrix = cv2.getRotationMatrix2D(center, -20, 1)  # Rotate 20 degrees right

        left_rotated = cv2.warpAffine(image, left_matrix, (image.shape[1], image.shape[0]))
        right_rotated = cv2.warpAffine(image, right_matrix, (image.shape[1], image.shape[0]))

        # Save the rotated images to the output folder
        base_name = os.path.splitext(os.path.basename(image_path))[0]
        left_image_path = os.path.join(output_folder, f"{base_name}_left.jpg")
        right_image_path = os.path.join(output_folder, f"{base_name}_right.jpg")

        cv2.imwrite(left_image_path, left_rotated)
        cv2.imwrite(right_image_path, right_rotated)

        print(f"Saved augmented images: {left_image_path}, {right_image_path}")

# Set up the directories
input_folder = r"C:\Users\Wiam\Desktop\AI security alarm system\LFW\lfw-deepfunneled\lfw-deepfunneled"
output_folder = r"C:\Users\Wiam\Desktop\AI security alarm system\augmented_images"

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Process each folder (person's images) inside the input directory
for person_folder in os.listdir(input_folder):
    person_folder_path = os.path.join(input_folder, person_folder)
    
    if os.path.isdir(person_folder_path):
        # Look for the front image (assuming it's the only image in the folder)
        for file in os.listdir(person_folder_path):
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                image_path = os.path.join(person_folder_path, file)
                augment_pose(image_path, output_folder)

print("Pose augmentation completed.")
```

```python
import cv2
import os
import random
import numpy as np
import dlib
from imutils import face_utils

def augment_data(image_path, output_folder):
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error loading image {image_path}")
        return

    # Load dlib's pre-trained face detector and predictor
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor(
        r"C:\Users\Wiam\Desktop\AI security alarm system\shape_predictor_68_face_landmarks.dat\shape_predictor_68_face_landmarks.dat"
    )

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)

    if len(faces) == 0:
        print(f"No face detected in {image_path}")
        return

    base_name = os.path.splitext(os.path.basename(image_path))[0]

    for face in faces:
        # Get the landmarks
        landmarks = predictor(gray, face)
        landmarks = face_utils.shape_to_np(landmarks)
        center = (int(landmarks[30][0]), int(landmarks[30][1]))  # Nose for rotation

        # Augmentations
        augmentations = []

        # 1. Horizontal Flip
        flipped = cv2.flip(image, 1)
        augmentations.append((flipped, f"{base_name}_flipped.jpg"))

        # 2. Brightness Adjustment
        bright_image = cv2.convertScaleAbs(image, alpha=1.2, beta=30)
        augmentations.append((bright_image, f"{base_name}_bright.jpg"))

        # 3. Gaussian Blur
        blurred = cv2.GaussianBlur(image, (5, 5), 0)
        augmentations.append((blurred, f"{base_name}_blurred.jpg"))

        # 4. Random Noise
        noisy = image + (30 * np.random.randn(*image.shape)).astype(np.uint8)
        noisy = np.clip(noisy, 0, 255)
        augmentations.append((noisy, f"{base_name}_noisy.jpg"))

        # 5. Saturation Adjustment
        hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
        hsv[..., 1] = cv2.add(hsv[..., 1], 40)  # Increase saturation
        saturated = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
        augmentations.append((saturated, f"{base_name}_saturated.jpg"))

        # 6. Extreme Rotations
        left_rotated = cv2.getRotationMatrix2D(center, 40, 1)  # Rotate 40° left
        left_extreme = cv2.warpAffine(image, left_rotated, (image.shape[1], image.shape[0]))
        right_rotated = cv2.getRotationMatrix2D(center, -40, 1)  # Rotate 40° right
        right_extreme = cv2.warpAffine(image, right_rotated, (image.shape[1], image.shape[0]))
        augmentations.extend([(left_extreme, f"{base_name}_left_extreme.jpg"),
                              (right_extreme, f"{base_name}_right_extreme.jpg")])

        # 7. Cropping/Zooming
        h, w, _ = image.shape
        cropped = image[int(h * 0.1):int(h * 0.9), int(w * 0.1):int(w * 0.9)]  # Crop 10% from all sides
        zoomed = cv2.resize(cropped, (w, h))  # Resize to original size
        augmentations.append((zoomed, f"{base_name}_zoomed.jpg"))

        # 8. Occlusions
        occluded = image.copy()
        h, w, _ = image.shape
        cv2.rectangle(occluded, (w // 3, h // 3), (w // 2, h // 2), (0, 0, 0), -1)  # Black rectangle on face
        augmentations.append((occluded, f"{base_name}_occluded.jpg"))

        # Save all augmented images in the output folder
        for aug_image, aug_name in augmentations:
            output_path = os.path.join(output_folder, aug_name)
            cv2.imwrite(output_path, aug_image)

        print(f"Augmented images saved for: {base_name}")


# Set up directories
input_folder = r"C:\Users\Wiam\Desktop\AI security alarm system\LFW\lfw-deepfunneled\lfw-deepfunneled"
output_base_folder = r"C:\Users\Wiam\Desktop\AI security alarm system\augmented_images"

if not os.path.exists(output_base_folder):
    os.makedirs(output_base_folder)

# Process each person's folder
for person_folder in os.listdir(input_folder):
    person_folder_path = os.path.join(input_folder, person_folder)
    person_output_folder = os.path.join(output_base_folder, person_folder)

    if not os.path.exists(person_output_folder):
        os.makedirs(person_output_folder)

    if os.path.isdir(person_folder_path):
        for file in os.listdir(person_folder_path):
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                image_path = os.path.join(person_folder_path, file)
                augment_data(image_path, person_output_folder)

print("Data augmentation completed!")
```

```python
import os
import shutil

def standardize_and_merge_folders(base_path):
    """
    Standardize folder names by matching them with filenames inside and merging duplicate folders.

    Args:
        base_path (str): Path to the parent directory containing the folders to validate and standardize.
    """
    # Dictionary to store the standardized folder names and their paths
    standardized_folders = {}
    mismatched_folders = []  # For logging folders with no matching files

    for folder_name in os.listdir(base_path):
        folder_path = os.path.join(base_path, folder_name)

        if os.path.isdir(folder_path):  # Ensure it's a folder
            standardized_name = None

            for file_name in os.listdir(folder_path):
                if file_name.lower().endswith(('.jpg', '.jpeg', '.png')):
                    # Extract the standardized name from the filename (e.g., 'FirstName_LastName')
                    standardized_name = "_".join(file_name.split("_")[:2])  
                    break

            if standardized_name:
                # Merge folders with the same standardized name
                if standardized_name in standardized_folders:
                    target_folder = standardized_folders[standardized_name]

                    # Move all files from the current folder to the target folder
                    for file in os.listdir(folder_path):
                        src_file = os.path.join(folder_path, file)
                        dest_file = os.path.join(target_folder, file)

                        # Prevent overwriting files with the same name
                        if os.path.exists(dest_file):
                            base, ext = os.path.splitext(file)
                            dest_file = os.path.join(target_folder, f"{base}_dup{ext}")

                        shutil.move(src_file, dest_file)

                    # Forcefully remove the non-empty folder
                    shutil.rmtree(folder_path)
                else:
                    # Rename the folder to match the standardized name if needed
                    new_folder_path = os.path.join(base_path, standardized_name)
                    if folder_path != new_folder_path:
                        shutil.move(folder_path, new_folder_path)

                    # Store the standardized name in the dictionary
                    standardized_folders[standardized_name] = new_folder_path
            else:
                mismatched_folders.append(folder_path)

    # Log folders that could not be matched
    if mismatched_folders:
        print("Folders with no matching files:")
        for folder in mismatched_folders:
            print(f" - {folder}")


if __name__ == "__main__":
    # Path to the parent directory containing the augmented folders
    parent_directory = r"C:\Users\Wiam\Desktop\AI security alarm system\augmented_images"

    # Call the function
    standardize_and_merge_folders(parent_directory)
    print("Folder standardization and merging completed!")
```

```python
import os
import shutil

def remove_small_folders(parent_folder, min_images=11):
    """
    Remove folders within the specified directory if they contain fewer than a specified number of images.
    
    Args:
        parent_folder (str): Path to the parent directory containing subfolders.
        min_images (int): Minimum number of images required to retain the folder.
    """
    for folder_name in os.listdir(parent_folder):
        folder_path = os.path.join(parent_folder, folder_name)
        if os.path.isdir(folder_path):
            # Count the number of image files in the folder
            image_count = sum(
                1 for file in os.listdir(folder_path) 
                if file.lower().endswith(('.png', '.jpg', '.jpeg', '.webp', '.avif'))
            )
            
            # Remove the folder if it has fewer than the required number of images
            if image_count < min_images:
                shutil.rmtree(folder_path)
                print(f"Removed folder: {folder_path} (contained {image_count} images)")

# Define the path to the augmented images folder
augmented_images_folder = r"C:\Users\Wiam\Desktop\AI security alarm system\augmented_images"

# Remove folders with fewer than 11 images
remove_small_folders(augmented_images_folder)
```
